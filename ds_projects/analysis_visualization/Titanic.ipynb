```python
df.drop(columns=['Cabin','Name'])
```

```python
df['Sex'].replace([0,1],['female','male'],inplace=True)
df.head()
```

```python
df['Sex'] = df['Sex'].replace({'male': 0, 'female': 1})
df = df.drop('Cabin', axis=1)
```

```python
embarked_list = df['Embarked'].tolist()
print(embarked_list)
```

```python
parch_list = df['Parch'].tolist()
print(parch_list)
```

```python
# Crear el grÃ¡fico de histograma
plt.figure(figsize=(8, 6))
sns.histplot(data=df, x='Embarked', discrete=True, shrink=0.8)

# AÃ±adir etiquetas y tÃ­tulo
plt.xlabel('Embarked')
plt.ylabel('Frequency')
plt.title('Frequency of Embarked Values')

# Mostrar el grÃ¡fico
plt.show()
```

```python
df = df.drop('Name', axis=1)
df.head()
```

```python
# Crear el grÃ¡fico de histograma
plt.figure(figsize=(8, 6))
sns.histplot(df['Age'], kde=False, bins=30)
# AÃ±adir etiquetas y tÃ­tulo
plt.title('Edades de los pasajeros')

# Mostrar el grÃ¡fico
plt.show()
```
